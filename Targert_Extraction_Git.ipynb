```python
# Bibliotecas 
import pandas as pd
import os
```

```python
os.chdir('root/')
file_names = os.listdir('root/Raw_Data_Targets/')

posicao_defeito_elementos = {}
for file_name in file_names:
  
    arquivo_rotulos = pd.read_excel('Raw_Data_Targets/' + file_name, names = ['Indice', 'Rotulo', 'Dist_Inicial', 'Dist_Final', 'Extensao'])
    aux = file_name.split('_')
    num_elem = aux[0]

    if (arquivo_rotulos['Rotulo'] == 'Trincado').any():
        indice_dormente = arquivo_rotulos.loc[arquivo_rotulos['Rotulo'] == 'Trincado', ['Indice']]
        indice_dormente = indice_dormente['Indice'].values[:]
        espacamento_dormente = (arquivo_rotulos.loc[1 , ['Extensao']]/len(arquivo_rotulos)).values[:]
        posicao_defeito = (indice_dormente*espacamento_dormente).tolist()
        rounded_posicao_defeito = [round(num) for num in posicao_defeito]        
       
        posicao_defeito_elementos[num_elem] = rounded_posicao_defeito
```
